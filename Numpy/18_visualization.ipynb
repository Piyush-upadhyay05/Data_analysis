# I use matplotlib to create visualizations of this twodimensional array:

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 1D array of points from -5 to 5 with step 0.01
points = np.arange(-5, 5, 0.01)

# Create a meshgrid: xs and ys will be 2D arrays of shape (1000, 1000)
xs, ys = np.meshgrid(points, points)

# Compute the distance from origin for each (x, y) pair
z = np.sqrt(xs**2 + ys**2)

# Plotting the result
plt.imshow(z, cmap=plt.cm.gray, extent=[-5, 5, -5, 5])  # Display 2D function as image
plt.colorbar()  # Add color scale on the side
plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid of values")  # Title with LaTeX-style math
plt.xlabel("x")
plt.ylabel("y")
plt.show()  # Show the plot window

```
